```python
import matplotlib.pyplot as plt
import numpy as np
# Define the quadratic function
def weather_model(x, a, b, c):
    return a * x**2 + b * x + c
# Read parameters from the file
with open('parameters.txt', 'r') as f:
    lines = f.readlines()
# Generate x values using NumPy linspace
x_values = np.linspace(-10, 10, 400)
# Create a figure for the plot
plt.figure(figsize=(10, 6))
# Loop through each line in the file to extract parameters
for line in lines:
    x_specific, a, b, c = map(float, line.split())

    # Calculate y values for the entire range of x values
    y_values = weather_model(x_values, a, b, c)
    # Calculate y value for the specific x value
    y_specific = weather_model(x_specific, a, b, c)
    # Plot the quadratic function for the entire range
    plt.plot(x_values, y_values, label=f'{a}x^2 + {b}x + {c}')
    # Plot the specific point in red
    plt.plot(x_specific, y_specific, 'ro')
# Adding labels and title to the plot
plt.title('Quadratic Functions')
plt.xlabel('x')
plt.ylabel('y')
# Displaying the legend
plt.legend()
# Displaying the grid on the plot
plt.grid(True)
# Showing the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    """
    Quadratic temperature model: T(time) = a*time^2 + b*time + c
    """
    return a * time**2 + b * time + c

def plot_temperature_model(a, b, c):
    """
    Plot the quadratic temperature model for a specific day.
    """
    time_values = np.linspace(0, 24, 100)  # 24 hours in a day
    temperature_values = quadratic_temperature_model(time_values, a, b, c)

    plt.figure(figsize=(10, 6))
    plt.plot(time_values, temperature_values, label='Temperature Model', color='blue')
    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature')
    plt.title('Quadratic Temperature Model for a Day')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    # Hard-coded coefficients for the quadratic equation
    a_fixed = 0.02
    b_fixed = 1.5
    c_fixed = 20

    # Get user input for coefficients (you can customize this part)
    a_user = float(input("Enter a coefficient for time^2: "))
    b_user = float(input("Enter a coefficient for time: "))
    c_user = float(input("Enter a constant term: "))

    # Combine fixed and user coefficients
    a_combined = a_fixed + a_user
    b_combined = b_fixed + b_user
    c_combined = c_fixed + c_user

    # Plot the temperature model
    plot_temperature_model(a_combined, b_combined, c_combined)

if __name__ == "__main__":
    main()
```
